```python
# Data Set Creation
import numpy as np
X_train= np.random.uniform(low=0, high=100, size=(50,))
y_train= 4*X_train+3
```

```python
# Determining of Yi
def ValeurYi(a,b,x):
  return a*x+b
```

```python
# Determining of RMSE
def RMSE(y_pred,y_train):
  n=y_train.shape[0]
  return (np.sqrt(np.sum((y_pred-y_train)*(y_pred-y_train)))*np.sqrt(1/n))
```

```python
# Foonction iteration
import numpy as np
def fonction_iteration (X_train,y_train,nb_iterations,LR,error_stop):
  i=0
  err=1000
  a=0.1
  b=0
  errorlist=[]
  n=X_train.shape[0]

  while (i < nb_iterations) and (err > error_stop) :
        
    a_gradient=(-np.sum(X_train*np.sqrt(np.abs(y_train-a*X_train-b))))*np.sqrt(1/n)
    b_gradient=(-np.sum(np.sqrt(np.abs(y_train-a*X_train-b))))*np.sqrt(1/n)
    
    a=a-a_gradient*LR
    b=b-b_gradient*LR
    
    y_pred=ValeurYi(a,b,X_train)
    err=RMSE(y_pred,y_train)
    errorlist.append(err)
    i+=1

    
    
    print('// Iteration num',i,' : a = ',a,',b = ',b,',err = ',err,'}')
  return a,b,errorlist
```

```python
error_list=[]
a,b,error_list=fonction_iteration(X_train,y_train,1600,0.000001,10e-5)
```
